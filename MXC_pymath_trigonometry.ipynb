# Variance

$\sigma^2 = \frac{\sum(x-\mu)^2}{N}$

## Standard Deviation

$\sigma = \sqrt\frac{\sum(x-\mu)^2}{N}$

```python
nums = np.random.randn(1000)

plt.subplot(121)
plt.plot(nums, 's')

plt.subplot(122)
plt.hist(nums)
plt.show()

print('This distribution has a mean of %s and a standard deviation of %s' % (np.mean(nums), np.std(nums)))
```

```python
mean = 15
std = 4.3
nums = np.random.randn(1000)*std + mean

plt.subplot(121)
plt.plot(nums, 's')

plt.subplot(122)
plt.hist(nums)
plt.show()

print('This distribution has a mean of %s and a standard deviation of %s' % (np.mean(nums), np.std(nums)))
```

```python
mean = 15
std = 4.3
nums = np.random.randn(1000)
nums = nums - np.mean(nums)
nums = nums/np.std(nums)

nums = nums*std + mean

plt.subplot(121)
plt.plot(nums, 's')

plt.subplot(122)
plt.hist(nums)
plt.show()

print('This distribution has a mean of %s and a standard deviation of %s' % (np.mean(nums), np.std(nums)))
```

```python
mean = 15
std = 4.3
nums = np.random.normal(mean, std, 1000)

plt.plot(nums, 's')
plt.show()

plt.hist(nums)
plt.show()
print('This distribution has a mean of %s and a standard deviation of %s' % (np.mean(nums), np.std(nums)))
```

```python
def angle_convert_plot(angle: float, unit: str, print_input=True):
  assert angle is not None and isinstance(angle, float) or isinstance(angle, int), f"angle must be float, not: '{angle}'"
  assert unit is not None and ("radian" == unit or "degree" == unit), f"unit must be radian or degree, not: '{unit}'"

  if print_input:
    print(f"What angle to convert? {angle}")
    print(f"Which unit did you input (radian or degree)? {unit}")

  if "degree" == unit:
    deg = angle
    rad = (deg*np.pi/180) % (2*np.pi)
  elif "radian" == unit:
    rad = angle
    deg = (rad*180/np.pi) % 360
  else:
    raise Exception("Invalida condition, must never happen")

  plt.plot([0, np.cos(rad)], [0, np.sin(rad)], 'r-', linewidth=3)
  plt.plot([0, 1], [0, 0], 'r-', linewidth=3)

  plt.title(f"Angle of {deg}, or {rad} rad")
  plt.axis('square')
  plt.axis([-1, 1, -1, 1])
  plt.grid()
  plt.show()

angle_convert_plot(np.pi/4, 'radian')
```

```python
def angle_convert_plot_dinamic():
  angle = eval(input("What angle to convert? "))
  unit = input("Which unit did you input (radian or degree)? ")
  angle_convert_plot(angle, unit, print_input=False)

angle_convert_plot_dinamic()
```

## The Pythagorean theorem

```python
a = 3
b = 4
c = np.sqrt(a**2 + b**2)

plt.plot([0, a], [0, 0], 'k', linewidth=2)
plt.plot([0, 0], [0, b], 'k')
plt.plot([0, a], [b, 0], 'k')

plt.plot([.3, .3], [0, .3], 'k', linewidth=1)
plt.plot([0, .3], [.3, .3], 'k', linewidth=1)

plt.text(a/3, -.5, f'a = {float(a)}', fontsize=15)
plt.text(.1, b/3, f'b = {float(b)}', fontsize=15)
plt.text(a/2+.1, b/2, f'c = {float(c)}', fontsize=15)

plt.axis('square')
plt.axis('off')
axlim = np.max((a, b)) + .5
plt.axis([-.5, axlim, -.5, axlim])
plt.show()
```

## Euler's formula

$e^{ik} = \cos(k) + i\sin(k)$

## Euler's formula with magnitude

$m*e^{ik} = m(\cos(k) + i\sin(k))$

```python
k = np.pi/4
m = 2.3

euler = m*np.exp(1j*k)
cis = m*(np.cos(k) + 1j*np.sin(k))

print(euler)
print(cis)
```

```python
mag = np.abs(euler)
ang = np.angle(euler)

print(m, mag)
print(k, ang)
```

```python
plt.polar([0, ang], [0, mag], 'r')
plt.polar(k, m, 'bo')
plt.show()
```

## Exercise: Euler's formula

```python
def euler_from_cos_sin():
  re = eval(input("cosine part: "))
  im = eval(input("sine part: "))

  m = np.sqrt(re**2 + im**2)
  k = np.arctan2(im, re)

  plt.polar([0, k], [0, m])
  plt.title('me$^{i\\phi}$, m=%g, $\\phi$= %g' % (m, k))
  plt.thetagrids([0, 90, 180, 270])
  plt.show()

euler_from_cos_sin()
```

## Exercise: random snakes from trigonometry

$x = cos(r_1\phi) \\
y = sin(r_2\phi) \\
r_1, r_2 = random(0, 1)$

```python
n = 1000
phi = np.linspace(0, 8*np.pi, n)
r1 = np.random.rand()
r2 = np.random.rand()

x = np.cos(r1*phi)
y = np.sin(r2*phi)

# connect one point with another
x = np.append(x, x[0])
y = np.append(y, y[0])

plt.plot(x, y)
plt.title(f"$r_1$ = {np.mean(r1):.2f}, $r_2$ = {np.mean(r2):.2f}")
plt.axis([-1.1, 1.1, -1.1, 1.1])
plt.axis('off')
plt.axis('square')
plt.show()
```
