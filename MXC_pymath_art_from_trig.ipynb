```python
import numpy as np
import matplotlib.pyplot as plt
```

## The asteroid radial curve

$x = a \ cos(t)^3 \\
y = a \ sin(t)^3 \\
a = t$

```python
t = np.linspace(-6*np.pi, np.pi*12, 1000)
a = t
x = a * np.cos(t)**3
y = a * np.sin(t)**3

plt.plot(x, y, 'm', linewidth=3)
plt.axis('square')
plt.axis('off')
plt.show()
```

## Exercise

$a = t^n \\
n \in \{0, \ 2, \ ..., \ 8\}$

```python
t = np.linspace(-6*np.pi, np.pi*12, 1000)

n = 0
for i in range(0, 3):
  for j in range(0, 3):
    a = t**n
    x = a * np.cos(t)**3
    y = a * np.sin(t)**3
    plt.subplot(int(f"{i+1}3{j+1}"))
    plt.plot(x, y, 'm')
    plt.axis('square')
    plt.axis('off')
    n += 1
  plt.show()
```

# Rose Curves

$x = cos(k \ t) \ cos(t) \\
y = cos(k \ t) \ sin(t)$

```python
k = 2
t = np.linspace(0, 4*np.pi, 1000)
x = np.cos(k*t) * np.cos(t)
y = np.cos(k*t) * np.sin(t)

plt.plot(x, y, 'm', linewidth=3)
plt.axis('square')
plt.axis('off')
plt.show()
```

# Squircle

$x^4+y^4=a^4$

```python
# x**4 + y**4 = a**4
# x**4 + y**4 = 1**4
# x**4 + y**4 = 1
# y**4 = 1 - x**4
# (y**4)**(1/4) = (1 - x**4)**(1/4)
# y = (1 - x**4)**(1/4)

x = np.linspace(-1, 1, 2001)
y = (1 - x**4)**(1/4)

plt.plot(x, y, 'k', linewidth=3)
plt.plot(x, -y, 'k', linewidth=3)
plt.axis('square')
plt.axis('off')
plt.show()
```

# Logaritmic spiral

$x = cos(t)e^{t/k} \\
y = sin(t)e^{t/k}$

```python
t = np.linspace(0, 18*np.pi, 1234)
k = -20
x = np.cos(t) * np.exp(t/k)
y = np.sin(t) * np.exp(t/k)

plt.plot(x, y, 'r')
plt.axis('square')
plt.axis('off')
plt.show()
```

```python
N = 150
ks = np.linspace(-6, -2, N)
t = np.linspace(0, 18*np.pi, 1234)

for i in range(N):
  x = np.cos(t) * np.exp(t/ks[i])
  y = np.sin(t) * np.exp(t/ks[i])
  plt.plot(x, y, color=[i/N, 0, i/N])
  plt.plot(x, -y, color=[0, 0, i/N])
  plt.plot(-x, y, color=[i/N, 0, 0])
  plt.plot(-x, -y, color=[i/N, i/N, i/N])

plt.axis('square')
plt.axis('off')
plt.show()
```

# Logistic map

$x_{n+1} = rx_{n}(1 - x_n)$

```python
rs = np.linspace(1, 4, 1000)
N = 500
x = 0.5 + np.zeros(N)
endcap = np.arange(round(N*0.9), N)

for ri in range(len(rs)):
  for n in range(N-1):
    x[n+1] = rs[ri]*x[n]*(1-x[n])
  u = np.unique(x[endcap])
  r = rs[ri]*np.ones(len(u))
  plt.plot(r, u, 'k.', markersize=1)

plt.show()
```
