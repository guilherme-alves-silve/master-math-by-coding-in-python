```python
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Math,display
```

Find integer solutions to $x^1+y^2=z^3$

```python
ints = np.arange(1,101)

z = np.zeros((len(ints),len(ints)))

# run the simulation
for x in ints:
    for y in ints:
        z[x-1, y-1] = (x**3 + y**3)**(1/3)

# check for integers
zInts = z%1 == 0

# now visualize
fig, ax = plt.subplots(1, 2, figsize=(12,6))
fig.tight_layout(pad=5)

# show the resulting z-values
h = ax[0].imshow(z)
fig.colorbar(h, ax=ax[0], fraction=.45)
ax[0].set_xlabel('y')
ax[0].set_ylabel('x')
ax[0].set_title('$z=\sqrt[3]{x^3+y^3}$', fontsize=16)

# show the boolean integer map
h = ax[1].imshow(zInts, vmin=0, vmax=1)
fig.colorbar(h,ax=ax[1], fraction=.045)
ax[1].set_xlabel('y')
ax[1].set_ylabel('x')
ax[1].set_title('Whether $z$ is int', fontsize=1)

plt.show()
```

```python
# print out those triplets

# run the simulation
for x in ints:
    for y in ints:
        z = (x**1 + y**2)**(1/3)
        
        # check for integers
        if z%1 == 0:
            display(Math('$%i^3 = %i + %i^2$' %(z, x, y)))
```

Compute and plot the following two sequences:

$\huge 
s_1 = \frac{-1^a}{\ln(\frac{1}{e^a})}\\
\huge
s_2 = \frac{\sqrt{a}}{\left(\frac{-\pi}{2}\right)^a}$

for $a \in \mathbb{Z}$

```python
nums = np.arange(1, 22)

s1 = np.zeros(len(nums))
s2 = np.zeros(len(nums))

for a in nums:
    s1[a-1] = (-1)**a / np.log(1/np.exp(a))
    s2[a-1] = np.sqrt(a) / (-np.pi/2)**a

fig, ax = plt.subplots(1, 2, figsize=(10,4))
ax[0].plot(s1)
ax[0].set_title('Sequence $s_1$')

ax[1].plot(s2)
ax[1].set_title('Sequence $s_2$')

plt.show()
```
