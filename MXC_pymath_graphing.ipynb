```python
# y = mx + b

x = [-5, 5]
m = 2
b = 1

y = [0, 0]
for i in range(0, len(x)):
  y[i] = m*x[i] + b

plt.plot(x, y, label='y = %sx + %s' % (m, b))
plt.grid()
plt.xlim(x)
plt.ylim(x)
plt.legend()

axis = plt.gca()
print(axis.get_xlim())
print(axis.get_ylim())

plt.plot(axis.get_xlim(), [0, 0], 'k--')
plt.plot([0, 0], axis.get_ylim(), 'k--')

plt.show()
```

```python
x, a = sym.symbols("x,a")

n = 1
expr = a/x

plot = symplot(expr.subs(a, n), (x, -5, 5), show=False)
plot[0].line_color = 'b'
plot[0].label = '$y = %s$' % sym.latex(expr.subs(a, n))

plot2 = symplot(expr.subs(a, 2), show=False)
plot2[0].line_color = 'r'
plot2[0].label = '$y = %s$' % sym.latex(expr.subs(a, 2))

plot.extend(plot2)
plot.ylim = [-5, 5]
plot.legend = True
plot.show()
```

```python
import random

# Exercise

# y = a/(x^2 - a)

x, a = sym.symbols("x,a")
a = [1, 2, 3, 4]
colors = ['blue', 'black', 'yellow', 'brown']

main_plot = None
for i, a in enumerate(a):
  y = a/(x**2 - a)
  actual_plot = symplot(y, (x, -5, 5), show=False)
  if main_plot:
    main_plot.extend(actual_plot)
  else:
    main_plot = actual_plot

  color = random.choice(colors)
  actual_plot[0].line_color = color
  colors.remove(color)

  actual_plot[0].label = '$%s$' % sym.latex(y)

main_plot.ylim = [-5, 5]
main_plot.legend = True
main_plot.show()
```

```python
# Exercise (onigiri)

#f(x) = -x^2

bridge = np.array([[x, (-x**2)] for x in np.linspace(-2, 2, 50)])
bridge_poly = Polygon(bridge, facecolor='g', edgecolor='k', alpha=.2)

square = np.array([[-.5, -4], [0.5, -4], [0.5, -2.5], [-.5, -2.5]])
square_poly = Polygon(square, facecolor='k')

_, ax = plt.subplots()
ax.add_patch(bridge_poly)
ax.add_patch(square_poly)
ax.set_ylim([-4.2, 0.2])
ax.set_xlim([-2.2, 2.2])
plt.show()
```

```python
x = np.linspace(-2, 2, 100)
f = -x**2

pnts = np.vstack((x, f)).T
poly = Polygon(pnts, facecolor='g', alpha=.2)

y = np.array([[-.5, -4], [-.5, -2.5], [.5, -2.5], [.5, -4]])
y_poly = Polygon(y, facecolor='k')

fig, ax = plt.subplots()
ax.add_patch(poly)
ax.add_patch(y_poly)

plt.plot(x, f, 'k')
plt.plot([-2, 2], [-4, -4], 'k')
ax.set_ylim([-4.2, 0.2])
ax.set_xlim([-2.2, 2.2])
plt.axis('off')
plt.show()
```
