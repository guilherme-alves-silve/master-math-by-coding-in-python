## Perfect number:

A positive integer that equals the sum of its positive divisors (excluding the number).

$
6 = 1 + 2 + 3 \\
8 \neq 1 + 2 + 4 \\
28 = 1 + 2 + 4 + 7 + 14
$

```python
import numpy as np
import sympy as sy
import matplotlib.pyplot as plt
```

```python
def is_perfect(number: int) -> bool:
  perfect = 0
  for i in range(1, number):
    if number % i == 0:
      perfect += i

  return perfect == number

print(is_perfect(6))
print(is_perfect(8))
print(is_perfect(28))
```

```python
perfect_nums = []
for number in range(2, 10_000+1):
  if is_perfect(number):
    perfect_nums.append(number)

print(perfect_nums)
```

## Pythagorean triplets:
Sets of integers that work.

Examples:
$
5^2 = 3^2 + 4^2 \\
13^2 = 5^2 + 12^2
$

Euclid's formula for finding triplets:

$
a = m^2 - n^2 \\
b = 2mn \\
c = m^2 + n^2 \\
m > n
$

```python
a = []
b = []
c = []
number_range = np.arange(2, 50+1)
for m in number_range:
  for n in number_range:
    a.append(m**2 - n**2)
    b.append(2*m*n)
    c.append(m**2 + n**2)
    

plt.scatter(a, b, c=c, marker='o', alpha=.5, linewidth=.5, edgecolors='k')
plt.axis('off')
plt.show()
```

# Fermat's last theorem

$
x^n + y^n = z^n
$

Has no integer solutions for n>2.

```python
ints = np.arange(1, 100+1)

Z = np.zeros((len(ints), len(ints)), dtype=int)
for x in ints:
  for y in ints:
    z = (x**3 + y**3)**(1/3)
    Z[x-1, y-1] = z

Zints = Z % 1 == 0
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].imshow(Z)
ax[0].set_xlabel("y")
ax[0].set_ylabel("x")
ax[0].set_title("$Z = \\sqrt[3]{x^3 + y^3}$", fontsize=16)

ax[1].imshow(Zints)
ax[1].set_xlabel("y")
ax[1].set_ylabel("x")
ax[1].set_title("Wheter $Z$ is an int font", fontsize=16)

plt.show()
```

## Heron's method

$
\sqrt{Y} = X \quad \Rightarrow \quad Y = X^2 \\
Y = (x + \epsilon)^2 \\
Y = (x + \epsilon)(x + \epsilon) \\
Y = x^2 + \epsilon^2 + 2x\epsilon \\
Y - x^2 = \epsilon^2 + 2x\epsilon \\
Y - x^2 = (\epsilon + 2x)\epsilon \\
\text{Ignore the } \epsilon \text{ in the left side because it's to small} \\
\frac{Y - x^2}{2x + \epsilon} = \frac{(2x + \epsilon)\epsilon}{\epsilon + 2x} \\
\frac{Y - x^2}{2x} \simeq \epsilon \\
X \simeq x + \frac{Y - x^2}{2x} \\
x \longleftarrow x + \frac{Y - x^2}{2x} \\
x \longleftarrow \frac{2x^2}{2x} + \frac{Y - x^2}{2x} \\
x \longleftarrow \frac{x^2 + Y}{2x} \\
x \longleftarrow \frac{x + Y / x}{2} \\
x \simeq X = \sqrt{Y}
$

```python
Y = 100

x = Y/3
for n in range(6):
  print(x)
  x = (x + Y/x) / 2

x
```

```python
nums2sqrt = np.linspace(2, 100+1, 50)
niterations = np.arange(3, 9)

err = np.zeros((len(niterations), len(nums2sqrt)))
for ni, num in enumerate(nums2sqrt):
  Y = num
  for ii, iters in enumerate(niterations):
    # Heron's algorithm
    x = Y/3 # initial guess
    for n in range(iters):
      x = (x + Y/x) / 2

    err[ii, ni] = abs(x-np.sqrt(Y))
```

```python
plt.title("Error thought iterations")
plt.imshow(-np.log(err), aspect=10, 
           extent=[nums2sqrt[0], nums2sqrt[-1], niterations[-1], niterations[0]])
plt.show()
```

## Exercise

Heron's mosquito spaceship

```python
Y = 13
start_estimative = np.linspace(-Y, Y, 40)
niters = 5

sqrt_results = np.zeros((len(start_estimative), niters+1))

for idx, start_num in enumerate(start_estimative):
  
  xs = [start_num]
  for n in range(niters):
    x = (xs[n] + Y/xs[n]) / 2
    xs.append(x)
  sqrt_results[idx,:] = xs

fig, _ = plt.subplots(1, figsize=(8, 6))
plt.plot(np.arange(niters+1), sqrt_results.T)
plt.title("Estimating $\\sqrt{13} = %f$" % x)
plt.xlabel("Algorithm iteration")
plt.ylabel("Numerical approximation")
plt.show()
```

## Smooth numbers

k-smooth number: An integer whose largest prime factor is $\le$ k

$
30 = 2 \times 3 \times 5  \\
30 \text{ is 5-smooth} \\
30 \text{ is 7-smooth} \\
30 \text{ is 11-smooth} \\
30 \text{ is 13-smooth} \\
...
$

```python
sy.primefactors(30)
```

```python
maxn = 10_000
largest_prime_factor = np.zeros(maxn+1, dtype=int)
for i in range(2, maxn+1):
  largest_prime_factor[i-2] = np.max(sy.primefactors(i))

plt.plot(largest_prime_factor, 'k.', markersize=.5)
plt.xlabel("Number")
plt.ylabel("Largest prime factor")
plt.show()
```

```python
unique_max_primes = np.unique(largest_prime_factor)

counts = np.zeros(len(unique_max_primes), dtype=int)
for idx, uni_num in enumerate(unique_max_primes):
  counts[idx] = np.sum(largest_prime_factor==uni_num)

plt.title(f"{unique_max_primes[np.argmax(counts)]} is the modal max prime factor")
plt.xlim([-2, 500])
plt.plot(unique_max_primes, counts, 'k.')
plt.show()
```
