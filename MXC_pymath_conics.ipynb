# Parabolas: formulas

## General formula
$4p(y - k) = (x - h)^2 \\
\text{h parameter} = \text{offset on x axis} \\
\text{k parameter} = \text{offset on y axis}$

## Reduced formula
$y = \frac{1}{4p}(x - h)^2 + k$

```python
# parameters

a = 1 # 1/4p
h = 1 # shift on x-axis
k = -2 # shift on y-axis
n = 1000
xlimit = 10

x = np.linspace(h-xlimit, h+xlimit, n)
y = a*(x - h)**2 + k

plt.grid()
plt.plot(x, y)
plt.show()
```

```python
plt.grid()
plt.plot(y, x)
plt.show()
```

```python
plt.grid()
plt.plot(-y, x)
plt.show()
```

```python
plt.grid()
plt.plot(x, -y)
plt.show()
```

```python
# Exercise

a = 3
p = 1/(4*a)
h = 2
k = 1
n = 1000
domain = 10

x = np.linspace(h-domain, h+domain, n)
y = a*(x - h)**2 + k

plt.plot(x, y, label='Parabola')
plt.axis([-6, 6, 0, 10])

# vertex: (h, k)

plt.plot(h, k, 'ro', label='vertex')

# focus: (h, k + p)

plt.plot(h, k + p, 'go', label='focus')

# directrix: y = k - p

plt.axhline(k - p, color='orange', label='directrix')

plt.axis([-3, 3, 0, 2])
plt.grid()
plt.legend()
plt.show()
```

# Circles: formulas

## General formula

$(x - x_1)^2 + (y - y_1)^2 = r^2$

## Reduced formula

$(x - x_1)^2 + (y - y_1)^2 - r^2 = 0$

```python
h = 2
k = -3
r = 3

axlim = r + max(abs(h), abs(k))
x = np.linspace(-axlim, axlim, 100)
X, Y = np.meshgrid(x, x)

Fxy = (Y-k)**2 + (X-h)**2 - r**2

plt.contour(X, Y, Fxy, 0)
plt.plot([-axlim, axlim], [0, 0], 'k--')
plt.plot([0, 0], [-axlim, axlim], 'k--')
plt.plot(h, k, 'go')
plt.gca().set_aspect('equal')
plt.show()
```

```python
scales = np.linspace(1, 2.5, 10)
params = [(-2, 0, 3), (2, 0, 3)]

for scale in scales:
  for h, k, r in params:
    h = h * scale
    k = h * scale
    r = r * scale
    axlim = scale * r + max(abs(h), abs(k))
    x = np.linspace(-axlim, axlim, 100)
    X, Y = np.meshgrid(x, x)

    Fxy = (Y-k)**2 + (X-h)**2 - r**2

    plt.contour(X, Y, Fxy, 0)
    plt.plot(h, k, 'ko')

plt.gca().set_aspect('equal')
plt.show()
```

```python
count = 15
centers = [-1.5, 1.5]
radius = np.linspace(.5, 3, count)
rgbs = np.linspace(0, 0.9, count)
k = 0

for r, color in zip(radius, rgbs):
  for center in centers:
    axlim = r + max(abs(center), abs(k))
    x = np.linspace(-axlim, axlim, 100)
    X, Y = np.meshgrid(x, x)

    Fxy = (Y-k)**2 + (X-center)**2 - r**2

    plt.contour(X, Y, Fxy, 0, colors=[(color, color, color)], linewidths=2)
plt.plot(centers, [k]*len(centers), color='k', linewidth=4)

plt.gca().set_aspect('equal')
plt.axis('off')
plt.show()
```

```python
h = [-1.5, 1.5]
x = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, x)

for r in np.linspace(.5, 3, 15):
  for hi in h:
    Fxy = (X-hi)**2 + Y**2 - r**2
    plt.contour(X, Y, Fxy, 0, colors=[r/3, r/3, r/3])

plt.axis('off')
plt.gca().set_aspect('equal')
plt.plot(h, [0, 0], 'k', linewidth=3)
plt.show()
```

# General formula of ellipse

$\frac{(x-x_1)^2}{a^2} + \frac{(y - y_1)^2}{b^2} = 1 \\
\text{center: } (x_1, y_1) \\
\text{radius: } a, b$

# Reduced formula of ellipse

$\frac{(x-x_1)^2}{a^2} + \frac{(y - y_1)^2}{b^2} - 1 = 0 \\
\text{center: } (x_1, y_1) \\
\text{radius: } a, b$

```python
x1 = 1
y1 = 2
a = 2
b = 3

axlim = np.max((a, b)) + np.max((abs(x1), abs(y1)))
x = np.linspace(-axlim, axlim, 100)
X, Y = np.meshgrid(x, x)

ellipse = (X-x1)**2/a**2 + (Y-y1)**2/b**2 - 1

plt.contour(X, Y, ellipse, 0)
plt.plot(x1, y1, 'go')
plt.gca().set_aspect('equal')
plt.show()
```

```python
# h is x1, k is y1

n = 16
x1 = 0
y1 = np.linspace(-4, 4, n)
a = abs(y1)
b = 4

x = np.linspace(-8, 8, 100)
X, Y = np.meshgrid(x, x)

for i in range(0, n):
  F = (X-x1)**2/a[i]**2 + (Y-y1[i])**2/b**2 - 1
  plt.contour(X, Y, F, 0, colors=[i/n, 0, i/n])

plt.axis('off')
plt.gca().set_aspect('equal')
plt.show()
```

# Hyperbolas: graphs and maths

## Reduced formula of hyperbolas

$\frac{(x - x_1)^2}{a^2} - \frac{(y-y_1)^2}{b^2} - 1 = 0 \\
\text{Center: } (x_1, y_1) \\
\text{Radii: } a, b$

```python
# h is x1 and k is y1

a = 1
b = .5
x1 = 1
y1 = 2

axlim = 2*np.max((a, b)) + np.max((abs(x1), abs(y1)))
x = np.linspace(-axlim, axlim, 100)
X, Y = np.meshgrid(x, x)

hyperbola = (X-x1)**2/a**2 - (Y-y1)**2/b**2 - 1

plt.contour(X, Y, hyperbola, 0)
plt.plot(x1, y1, 'go')
plt.plot([-axlim, axlim], [0, 0], 'k--')
plt.plot([0, 0], [-axlim, axlim], 'k--')

plt.gca().set_aspect('equal')
plt.show()
```

```python
# h is x1 and k is y1

a = 1
b = .5
x1 = 1
y1 = 2

axlim = 2*np.max((a, b)) + np.max((abs(x1), abs(y1)))
x = np.linspace(-axlim, axlim, 100)
X, Y = np.meshgrid(x, x)

hyperbola = (Y-y1)**2/b**2 - (X-x1)**2/a**2 - 1

plt.contour(X, Y, hyperbola, 0)
plt.plot(x1, y1, 'go')

plt.gca().set_aspect('equal')
plt.show()
```

```python
# h is x1 and k is y1

n = 16
x1 = 0
y1 = 0
axlim = 8
x = np.linspace(-axlim, axlim, 100)
X, Y = np.meshgrid(x, x)

ab = np.linspace(1, 5, n)

for i in range(0, n):

  FX1 = (X)**2/ab[i]**2
  FY2 = (Y)**2/ab[i]**2

  hyperbola_up_down = FX1 - FY2 - 1
  hyperbola_sides = FY2 - FX1 - 1

  plt.contour(X, Y, hyperbola_up_down, 0, colors=[i/n, 0, i/n])
  plt.contour(X, Y, hyperbola_sides, 0, colors=[0, i/n, i/n])

plt.gca().set_aspect('equal')
plt.axis('off')
plt.show()
```
